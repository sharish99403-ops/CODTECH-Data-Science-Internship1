Task 1 – Data Pipeline Development

Problem Statement
Develop an ETL (Extract, Transform, Load) pipeline using Python. 
The pipeline extracts data from a CSV file, performs preprocessing 
such as handling missing values, scaling numerical features, and 
encoding categorical variables, and finally loads the processed 
data into a new CSV file.


```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
# Extract
print("Loading dataset...")
df = pd.read_csv("data.csv")
print("Dataset Loaded Successfully")
df.head()
```

```python
# Identify columns
numerical_features = df.select_dtypes(include=["int64", "float64"]).columns
categorical_features = df.select_dtypes(include=["object"]).columns

# Numerical Pipeline
num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="mean")),
    ("scaler", StandardScaler())
])

# Categorical Pipeline
cat_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])

# Combine pipelines
preprocessor = ColumnTransformer([
    ("num", num_pipeline, numerical_features),
    ("cat", cat_pipeline, categorical_features)
])

# Apply transformation
print("Applying Transformations...")
processed_data = preprocessor.fit_transform(df)
print("Transformation Completed")
```

```python
# Convert to DataFrame
processed_df = pd.DataFrame(processed_data)

# Save file
processed_df.to_csv("processed_data.csv", index=False)

print("Processed data saved successfully!")
```

Conclusion

The ETL pipeline was successfully developed using Python. 
Data was extracted from the CSV file, transformed using 
preprocessing techniques, and loaded into a new processed file.
